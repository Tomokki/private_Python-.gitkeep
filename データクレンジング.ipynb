「売上履歴」に欠損値が含まれているか確認しなさい

```python
uriage_data.isnull().any(axis=0)
```

##金額の欠損値の補完方法  
集計期間中に商品単価の変動はないという前提条件がある  
つまり、欠損値は他の同じ商品の単価をもとに補完ができる  
複雑な処理になるため順を追って説明

```python
# item_price の中で欠損値のある個所を特定。この処理を実行する事
# で flg_is_null 変数にどの行に欠損値が存在するかというフラグを保持
flg_is_null = uriage_data["item_price"].isnull()

# ループ条件はデータが欠損している商品名の一覧
# 一番大枠の list() は変数の値をリスト形式に変換する処理
# .loc 関数は条件を付与し、それに合致するデータを抽出する事ができる
# 「金額が欠損している」行の商品名を抽出
# 「金額が欠損している」という条件付のために flg_is_null を渡している
# 2 番目の添字は条件に合致したデータの、どの列を取得するかを指定
# item_name とすることで欠損値の存在する「商品名」を抽出
# 最後の unique() は抽出した商品名の重複をなくし、一意にする命令
# 欠損値のある商品名は 1 回分かればよいので無駄なループ処理をなくす


for trg in list(uriage_data.loc[flg_is_null,"item_name"].unique()):
  price = uriage_data.loc[(~flg_is_null)
  &(uriage_data["item_name"]== trg),"item_price"].max()
# ループ変数である「欠損値がある商品名」を用いて、同じ商品で金額が
# 正しく記載されている行を .loc で探し、その金額を取得
# flg_is_null の「 」は否定演算子
# 欠損値がある商品と同じ商品データから金額を取得する事ができる


# 売上履歴の item_price 列に対して .loc を行い、欠損を起こしている
# 対象データを抽出し、先ほど生成した price を欠損値に代入
  uriage_data['item_price'].loc[(flg_is_null)
  &(uriage_data['item_name']==trg)] = price

# 金額の欠損値の補完結果を確認
uriage_data.head()
```

```python
uriage_data.isnull().any(axis=0)
```

####  
「売上履歴」の中の補完した金額の値の検証  
各商品名の金額の最大値と最小値を表示  
同一商品の金額の最大値と最小値が同じ値でなければ不整合

```python
# 1 行目ですべての商品に対してループ処理を実施
# ループの中で商品に設定されている金額の最大額と最小額を出力する処理を実行
for trg in list(uriage_data["item_name"].sort_values().unique()):
  print(trg + "の最大額:" + str(uriage_data.loc[uriage_data["item_name"]==trg]
                            ["item_price"].max())+"の最小額:" +
        str(uriage_data.loc[uriage_data["item_name"]==trg]
            ["item_price"].min(skipna=False)))

#  . skipna = における「 skipna 」は NaN データを無視するかに関する設定
# False を指定すると、 NaN が存在する場合、最小値は NaN と表示される
```

```python
# すべての商品の最大値と最小値が一致している事が確認できるので、金額の補完は成功している事が分かる
```

##練習問題⑦：日付の揺れの補正
「顧客台帳」の登録日を表示しなさい  
「数値」として取り込まれている登録日のデータの件数を確認しなさい  
「数値」として取り込まれている登録日のデータを日付型に変換  
「文字」として取り込まれている登録日のデータを日付型に変換  
数値から日付型に補正したデータと、文字から日付型に補正したデータを結合して、登録日のデータを更新しなさい  
登録日から登録月を算出し、顧客の数を集計しなさい


```python
# 「顧客台帳」の登録日を表示
print(kokyaku_data["登録日"].head())

# 数値として取り込まれているデータがみられる
```

```python
# 「数値」として取り込まれている登録日のデータの件数を確認しなさい
# 1 行目は、顧客台帳の登録日が数値かどうかを str.isdigit で判定

flg_is_serial = kokyaku_data["登録日"].astype("str").str.isdigit()


# str.isdigit の実行結果が True となっている対象件数を表示
# このデータは 22 件の登録日が数値として取り込まれていることがわかる

flg_is_serial.sum()
```

```python
# 「数値」として取り込まれている登録日のデータを日付型に変換
# pd.to_timedelta 関数を用いて、数値（シリアル値）から日付に変換

# Excel と Python の日付データのずれを解消するには以下のようにする

serial = kokyaku_data.loc[flg_is_serial,"登録日"].astype("float")
fromSerial = pd.to_timedelta(serial,unit="D") + pd.to_datetime("1900/01/01")
fromSerial

# Excel 上の日付と 2 日ずれて変換されている
# システムや言語によってシリアル値の扱い方の定義や、日付等の仕様に違いがある事が原因
```

```python
# Excel と Python で以下の仕様が相違
# Excel の仕様
# 最小日付は 1900/01/01 だが、シリアル値は 「 0 」 ではなく 「 1 」 から開始
# 本来うるう年ではない 1900/02/29 を有効な日付としてシリアル値に加算する
# Python の仕様
# 最小日付は 1900/01/01 だが、シリアル値は 「 1 」 ではなく 「 0 」 から開始
# 本来うるう年ではない 1900/02/29 を無効な日付としてシリアル値に加算しない
```

```python
# Excel と Python の日付データのずれを解消するには以下のようにする

serial = kokyaku_data.loc[flg_is_serial,"登録日"].astype("float")
fromSerial = pd.to_timedelta(serial -2 ,unit="D") + pd.to_datetime("1900/01/01")
fromSerial
```

```python
# 「文字」として取り込まれている登録日のデータを日付型に変換
# スラッシュ区切りの書式を、ハイフン区切りに統一するために処理

fromString = pd.to_datetime(kokyaku_data.loc[~flg_is_serial,"登録日"])
fromString
```

```python
# 数値から日付型に補正したデータと、文字から日付型に補正したデータを結合して、登録日のデータを更新しなさい

kokyaku_data["登録日"] = pd.concat([fromSerial,fromString])
kokyaku_data
```

```python
# 登録日から登録月を算出し、顧客の数を集計しなさい
kokyaku_data["登録年月"] = kokyaku_data["登録日"].dt.strftime("%Y%m")
rslt = kokyaku_data.groupby("登録年月").count()["顧客名"]
print(rslt)
print(len(kokyaku_data))
```
